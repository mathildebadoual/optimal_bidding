```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df_5min = pd.read_csv('consolidated_data/June2018_5min.csv')
df_30min = pd.read_csv('consolidated_data/June2018_30min.csv')
df_5min
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_30min['Timestamp'], y=df_30min['Price'], name='30min', line_color='#9e2b56'))
fig.update_layout(title_text="Electricity Prices", xaxis_rangeslider_visible=True,
    xaxis=dict(
        title='Time',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#4f6a79'
        )
    ),
    yaxis=dict(
        title='Electricity Price ($/MW)',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#4f6a79'
        )
    ))
fig.show()
```
