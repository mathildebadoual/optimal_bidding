```python
d = pd.read_csv('result_optimization_alone.csv')
d.set_index('time_step')
```

```python
plt.figure(figsize=(15, 8))
plt.plot(range(100, 300), d['power_cleared'].values[100:300], label='power_cleared')
plt.plot(range(100, 300), d['power_bid'].values[100:300], label='power_bid')
plt.plot(range(100, 300), 10 * d['price_bid'].values[100:300], label='price_bid')
plt.legend()
plt.xticks(range(100, 300, 4), d['time_step'].values[100:300:4], rotation=45)
plt.title('Bids and Clearance')
plt.xlabel('time')

plt.grid()
```

```python
plt.figure(figsize=(15, 8))
plt.plot(range(120, 220), d['power_cleared'].values[120:220], label='power_cleared')
plt.plot(range(120, 220), d['soe'].values[120:220], label='soe')
plt.plot(range(120, 220), 50 * d['price_bid'].values[120:220], label='price_bid')
plt.plot(range(120, 220), 50 * d['price_cleared'].values[120:220], label='price_bid')
plt.plot(range(120, 220), d['power_bid'].values[120:220], label='price_bid')
plt.legend()
plt.xticks(range(120, 220, 4), d['time_step'].values[120:220], rotation=45)
plt.title('SOE, Power and Price')
plt.xlabel('time')
plt.grid()
```

```python
plt.figure(figsize=(15, 8))
plt.plot(range(500, 700), d['power_cleared'].values[500:700], label='power_cleared')
plt.plot(range(500, 700), d['power_bid'].values[500:700], label='power_bid')
plt.plot(range(500, 700), 10 * d['price_bid'].values[500:700], label='price_bid')
plt.legend()
plt.xticks(range(500, 700, 4), d['time_step'].values[500:700:4], rotation=45)
plt.title('Bids and Clearance')
plt.xlabel('time')
plt.grid()
```
